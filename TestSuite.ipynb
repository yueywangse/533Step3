```python
import unittest
from canvasacademicinsights.tests.test_canvasinsights import TestMain
from canvasacademicinsights.tests.test_canvasdataretriever import TestCleaner
from canvasacademicinsights.tests.test_canvasaimodel import TestAIModel
from canvasacademicinsights.tests.test_canvasaiorganize import TestAIPrompt
from canvasacademicinsights.tests.test_canvassummaryvisualizer import TestSummaryVisuals
from canvasacademicinsights.tests.test_canvastimevisualizer import TestTimeVisuals


def my_suite():
    suite = unittest.TestSuite()
    result = unittest.TestResult()
    suite.addTest(TestMain('test_clean_data'))
    suite.addTest(TestMain('test_get_courses'))
    suite.addTest(TestCleaner('test_clean_data'))
    suite.addTest(TestCleaner('test_grade'))
    suite.addTest(TestCleaner('test_assignment_grade'))
    suite.addTest(TestCleaner('test_quiz_grade'))
    
    # Visuals
    suite.addTest(TestSummaryVisuals('test_plot_overall_scores'))
    suite.addTest(TestSummaryVisuals('test_plot_grade_distribution'))
    suite.addTest(TestSummaryVisuals('test_plot_missing_assignments'))
    suite.addTest(TestSummaryVisuals('test_plot_num_assignments_per_course'))
    
    suite.addTest(TestTimeVisuals('test_get_last_graded_items_normal'))
    suite.addTest(TestTimeVisuals('test_plot_weekly_average_score_heatmap'))

    # AI Suite this one takes a while to run
    suite.addTest(TestAIModel('testCleanData'))
    suite.addTest(TestAIModel('testAIModel'))
    suite.addTest(TestAIPrompt('testGrades'))
    suite.addTest(TestAIPrompt('testOrganize'))

    runner = unittest.TextTestRunner()
    print(runner.run(suite))
                      
my_suite()
```
