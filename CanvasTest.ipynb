```python
from canvasacademicinsights.canvasmain import canvasinsights as cai
data = cai.start()
```

```python
data
```

```python
for course in data:
    courseid = course[0]
    coursename = course[1]
    courseenrollment = course[2]
    print(courseid, coursename, courseenrollment)
    gradelist = course[3]
    for item in gradelist:
        # Note that some due dates are none, dates are stored as datetime object
        if item.isAssignment():
            print('Assignment',item.name, item.score, item.total, item.date)
        elif item.isQuiz(): # can be else too
            print('Quiz',item.name, item.score, item.total, item.date)
```

```python
from canvasacademicinsights.canvasmain import canvasdataretriever as c 
from datetime import datetime
import random 

courses = []

for j in range(5): 
    gradeList = []
    sampleCourse = [
        "courseid",
        "coursename",
        "courseenrollment",
    ]
    
    counter = 1
    for i in range(10): 
        counter += 1
        if random.random() > 0.5:
            gradeList.append(c.QuizGrade(name = f"Test{random.random()}", score = random.random() * 100, total = 100, date = datetime.strptime(f"2025-11-{counter}", "%Y-%m-%d")))
        else:
            gradeList.append(c.AssignmentGrade(name = f"Test{random.random()}", score = random.random() * 100, total = 100, date = datetime.strptime(f"2025-11-{counter}", "%Y-%m-%d")))
    sampleCourse.append(gradeList)
    print(gradeList)
    courses.append(sampleCourse)

print(courses)
from canvasacademicinsights.canvasvisualizer.canvastimevisualization import CanvasTimeVisualization

time_viz = CanvasTimeVisualization(courses)
time_viz.plot_item_scores(0)
time_viz.plot_all_last_graded_items_bar()
time_viz.plot_weekly_average_score_heatmap()
```
